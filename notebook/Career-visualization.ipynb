### Salaire moyen, salaire médian et déciles

En ce qui concerne les salaires, j'ai décidé de commencer par travailler sur les salaires annuels plutôt qu'horaires. 
J'ai décidé de présenter les déciles sous forme de tableau, ce qui me paraissait être la forme la plus simple et la plus claire. Je vais donc créer 2 colonnes à associer à ma DataFrame. DecileRank contenant le numéro du décile associé à la ligne. DecileValue contenant l'intervalle du décile associé à la ligne.
J'extrais ces valeurs dans des listes et j'en fais un tableau que j'affiche.
J'affiche également la médiane (qui est la valeur entre le 5ème et le 6ème décile).
Je finis par calculer la moyenne en sommant tous les salaires de ma table et en divisant par le nombre de lignes (tout en gardant 2 chiffres significatifs pour la lisibilité). 

```python
# Mise en place d'un format pour l'affichage des salaires
fmt='${x:,.0f}'
```

```python
# Create column giving the decile rank 
dfb['DecileRank'] = pd.qcut(dfb['SalaryUSD'], q = 10, labels = False) 
# Create column giving the decile range
dfb['DecileValue'] = pd.qcut(dfb['SalaryUSD'], q = 10) 

# Put the decile rank and values in lists
lrank = dfb.sort_values("DecileRank").DecileRank.unique()+1
lvalue = dfb.sort_values("DecileValue").DecileValue.unique()
# Create a new dataframe containing the decile rank and the range for this rank
dfdeciles = pd.DataFrame()
dfdeciles['Décile']=lrank
dfdeciles['Intervalle']=lvalue

print(dfdeciles)
print("\nLa médiane des salaires annuels est de ${:,.0f}.".format(lvalue[5].right))
print("Le salaire moyen annuel est de ${:,.0f}.".format(round(dfb['SalaryUSD'].mean(),2)))



```

### Salaire moyen en fonction du JobTitle

On va commencer par calculer le salaire moyen en groupant par JobTitle. On les trie par salaire moyen croissant pour rendre le graphique plus clair. Et on utilise un diagramme à barre qui est le plus approprié pour comparer des quantités entre différents groupes.

```python
seriesj = dfb.groupby('JobTitle').SalaryUSD.mean().sort_values()

# create a color palette the length of the dataframe
colors = sns.color_palette('husl', n_colors=len(seriesj.index))

ax=sns.barplot(x=seriesj.index, y=seriesj.values, palette=colors)

fig = plt.gcf()
# Modifie la taille du graphique
fig.set_size_inches(10,4)
# Ajout du titre
fig.suptitle("Salaire moyen en fonction du poste", fontsize=25)
# Suppresion du label pour l'axe x et ajout du label pour l'axe y
plt.xlabel("");
plt.ylabel("Salaire moyen", fontsize=20, rotation=90)
# Suppression des ticks de l'abscisse
ax.tick_params(axis='x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    labelbottom=False)

# Change the format of salary
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)

# create color map with colors and df.names
cmap = dict(zip(seriesj.index, colors))
# create the rectangles for the legend
patches = [Patch(color=v, label=k) for k, v in cmap.items()]
# add the legend
plt.legend(handles=patches, bbox_to_anchor=(1.04, 0.5), loc='center left', borderaxespad=0);
```

Pour le diagramme circulaire, on compte le nombre d'occurences de chaque JobTitle et on en fait un diagramme circulaire en affichant les pourcentages pour le rendre plus clair. 

```python
data = dfb.groupby("JobTitle").JobTitle.count().sort_values()
data.plot.pie(autopct="%.1f%%", radius=3);
```

### Est-il préférable de travailler pour une ou plusieurs entreprises ? 

Pour répondre à cette question, je choisis d'utiliser un histogramme avec barplot en visualisant le salaire moyen en fonction du nombre d'entreprises dans lesquelles les personnes travaillent. On les affiche en fonction du nombre d'entreprises pour lesquelles les personnes travaillent pour rendre le graphique plus lisible.

```python
seriesn = dfb.groupby('HowManyCompanies').SalaryUSD.mean()

newindex=[]
# Permet de réduier la longueur du texte en abscisse
for i in seriesn.index:
    if len(i)>10:
        i=i[0]
    newindex=newindex+[i]
    
# On fait notre diagramme a barres
ax=sns.barplot(x=newindex,y=seriesn.values)
fig = plt.gcf()
# Modifie la taille du graphique
fig.set_size_inches(11,8)
# Ajout du titre
fig.suptitle("Salaire moyen en fonction du nombre d'entreprises", fontsize=25)
# Ajout des labels pour les axes x et y
plt.xlabel("Nombre d'entreprises dans lesquelles les personnes travaillent", fontsize=20);
plt.ylabel("Salaire moyen", fontsize=20, rotation=90)
# Modification de la taille des valeurs des ticks pour la lisibilité
ax.tick_params(axis = 'both', labelsize = 16)
# Change the format of salary
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)
```

On voit que les personnes qui travaillent pour plusieurs entreprises gagnent, en moyenne, davantage annuellement que celles qui travaillent pour moins d'entreprises. On peut en déduire que d'un point de vue financier, il est préférable de travailler pour plusieurs entreprises.

### Le nombre d'années d'expérience dans ce type d'emploi a t il une influence sur le salaire ? 

On va utiliser un histogramme pour les mêmes raisons que précédemment. On va afficher le salaire moyen en fonction du nombre d'années d'expérience.

Le résultat n'étant pas très clair du fait du grand nombre d'abscisses différentes. Je choisis d'utiliser un nuage de points avec régression linéaire à l'aide de regplot.

Nous pouvons déduire de la courbe ascendante de notre régression linéaire que le salaire annuel a tendance à augmenter avec les années d'expérience. 

### Y-a-t-il une inégalité salariale entre les hommes et les femmes dans les emplois liés à la data aux Etats Unis ?

Nous voyons que les hommes sont en moyenne mieux payés que les femmes dans les emplois liés à la data aux Etats-Unis.

### Si cette inégalité salariale existe, est-elle due au fait que les femmes soient moins bien représentées dans les jobs les mieux payés ou que pour un travail identique elles sont en moyenne moins bien payées?

Commençons par regarder les disparités de salaires par genre pour chacun des JobTitle. 

```python
seriesjg = dfb.groupby(['JobTitle', 'Gender']).SalaryUSD.mean()
ax=seriesjg.unstack(level=0).plot(kind='bar', subplots=True, figsize=(8,25), legend=None, ylabel='Salaire moyen')
# Ajout du label Genre pour l'axe des abscisses
plt.xlabel("Genre", fontsize=15)
# Ajoute de l'espace entre les graphes
plt.tight_layout()
plt.figure(figsize=(150,150));
# Change format for each subplot
for axe in ax.ravel():
    # Change the format of salary
    tick = mtick.StrMethodFormatter(fmt)
    axe.yaxis.set_major_formatter(tick)
```

On peut en déduire que pour tous les métiers, les femmes sont moins bien payées que les hommes (sauf une exception pour les Data Scientist qui représentent une très faible partie des métiers concernés : 0.6% cf pie chart). On peut déjà en déduire que pour un travail identique, elles sont moins bien payées. 
Maintenant regardons comment les femmes sont réparties par niveau de salaire par rapport aux hommes en superposant les diagrammes de densité.

```python
# Extrait les données liées aux hommes
datam=dfb[dfb.Gender.isin(['Male'])]
sns.kdeplot(data=datam, x='SalaryUSD', cumulative=True)
# Extrait les données liées aux femmes
dataf=dfb[dfb.Gender.isin(['Female'])]
ax=sns.kdeplot(data=dataf, x='SalaryUSD', cumulative=True)

# On met une limite pour améliorer la lisibilité
plt.xlim(0,300000)
fig = plt.gcf()
# Modifie la taille du graphique
fig.set_size_inches(10,4)
# Ajout du titre
fig.suptitle("Densité par genre en fonction du salaire", fontsize=25)
# Ajout des labels pour les axes x et y
plt.xlabel("Salaire moyen", fontsize=20);
plt.ylabel("Densité", fontsize=20, rotation=90)
# Modification de la taille des valeurs des ticks pour la lisibilité
ax.tick_params(axis = 'both', labelsize = 16)
# Ajout d'une légende
plt.legend(["Homme","Femme"]);
# Change the format of salary
tick = mtick.StrMethodFormatter(fmt)
ax.xaxis.set_major_formatter(tick)
```

On constate que la courbe de densité de salaire des femmes en orange est au-dessus de celles des hommes en bleu. Cela signifie qu'il y a davantage de femmes qui ont des bas salaires que des hommes. On peut donc en conclure que les femmes sont moins bien représentées que les hommes dans les postes les mieux payés.

### Résumer en quelques lignes votre plan de carrière pour conquérir le rêve américain

Je choisis de visualiser le salaire en fonction du nombre d'heures travaillées par semaine. 

```python
data=dfb
# On transforme les données de HoursWorkerPerWeek pour ne garder que les nombres
data.HoursWorkedPerWeek = data.HoursWorkedPerWeek.astype(str)
data=data[data.HoursWorkedPerWeek.str.isdigit()]
# On transforme les données de HoursWorkedPerWeek en float pour pouvoir faire la régression linéaire
ax=sns.regplot(x=data.HoursWorkedPerWeek.astype(float), y=data.SalaryUSD, scatter_kws={"color": "black"}, line_kws={"color": "red"})

fig = plt.gcf()
# Modifie la taille du graphique
fig.set_size_inches(10,4)
# Ajout du titre
fig.suptitle("Salaire en fonction du nombre d'heure travaillées", fontsize=25)
# Ajout des labels pour les axes x et y
plt.xlabel("Heures de travail par semaine", fontsize=20);
plt.ylabel("Salaire moyen", fontsize=20, rotation=90)
# Modification de la taille des valeurs des ticks pour la lisibilité
ax.tick_params(axis = 'both', labelsize = 16)
# Change the format of salary
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)
```

Ce graphique confirme donc (comme on pouvait s'y attendre) que le salaire augmente avec le nombre d'heures travaillées.
